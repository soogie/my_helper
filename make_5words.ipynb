```python
import pandas as pd
```

```python
df_all = pd.read_csv("data/unigram_freq.csv")
```

```python
five_words = df_all[[len(w)==5 if not pd.isna(w) else False for w in df_all["word"]]]["word"]
```

```python
chars = [[], [], [], [], []]
for i in range(5):
    chars[i] = [w[i] for w in five_words]

df_5words = pd.DataFrame({"word": five_words, "char1": chars[0], "char2":chars[1], "char3":chars[2], "chars4":chars[3], "chars5":chars[4]})


    
```

```python
df_5words.to_pickle("data/fivewords.pkl")
```

```python
def word_check(word, querylist):
    for query in querylist:
        if not eval(query):
            return False
    return True
            
```

```python
ql = []
ql.append("word[0]=='a'")
ql.append("word[3]== 'c'")
ql.append("'o' not in word")
```
